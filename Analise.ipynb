## Carregando o conjunto de dados

```python
import pandas as pd

had = pd.read_excel ('datassets/HAD_BRASIL_REGIOES_UFS_2024.xlsx', skiprows=8)
                    
had
```

```python
had.info()
```

## Preparando os dados para a análise

**Obs.:** Todas as colunas estão sendo identificadas como Object

Antes de resolver esses problemas vamos selecionar as colunas Unidade Geográfica, Localização, Dependência Administrativa, Total Infantil, Total Fundamental e Total Médio para efetuar a análise.

```python
had = had.iloc[:, [1,2,3,4,7,20]]
had
```

Dicionário:
- ug = Unidade Geográfica
- loc = Localização
- da = Dependência Administrativa
- ti = Total Ensino Infantil
- tf = Total Ensino Fundamental
- tm = Total Ensino Médio

```python
colunas = ["ug", "loc", "da", "ti", "tf", "tm"]
had.columns = colunas
had
```

Vamos considerar que os valores ausentes indicam que não existem escolas com aquele nível de ensino na região.

```python
df = had.replace('--', 0)
```

```python
df.info()
```

```python
df.groupby("ug")[["ti", "tf", "tm"]].median().round(2)
```

```python
df.groupby("loc")[["ti", "tf", "tm"]].median()
```

```python
df.groupby(["ug", "loc"])[["ti", "tf", "tm"]].median()
```

```python
pd.pivot_table(df, values="tf", index="ug", columns="da", aggfunc="median")
```

## Valores não Nulos
Podemos criar um novo conjunto só com os valores não nulos.

```python
not_null = df[(df['ti']>0)&(df['tf']>0)&(df['tm']>0)]
```

Mas aqui temos alguns problemas extras, por exemplo a quantidade de linhas que são perdidas



```python
print("Linhas Antes:", len(df))
print("Linhas Depois:", len(not_null))
print("Perdas de Linhas:", len(df) - len(not_null))
```

Uma saída seria trocar o 0 por NAN ('not a number'). Pois o método de calcular médias vai ignorar esses valores NaN.

```python
import numpy as np

df_sem_zeros = df[["ug", "ti", "tf", "tm"]].copy()
df_sem_zeros[["ti", "tf", "tm"]] = df_sem_zeros[["ti", "tf", "tm"]].replace(0, np.nan)
```

```python
print("Média sem zeros: ", df_sem_zeros['ti'].mean().round(2))
print("Média com zeros: ", df['ti'].mean().round(2))
```

Mas aqui vale a pena ressaltar que existem municípios que não possuem escolas de alguns níveis. Então, Ambas são análises válidas, mas servem a propósitos diferentes:

## Visualização de dados

```python
import matplotlib.pyplot as plt

df.groupby("da")[["ti", "tf", "tm"]].mean().plot(kind="barh", figsize=(10, 5))
plt.title("Média de alunos por Dependência Administrativa")
plt.xlabel("Quantidade média de matrículas")
plt.ylabel("Tipo de administração")
plt.tight_layout()
plt.show()
```

```python
df.groupby("loc")[["ti", "tf", "tm"]].mean().plot(kind="bar", figsize=(8, 5))
plt.title("Média de matrículas por Localização")
plt.ylabel("Média de alunos")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
df.groupby("da")["tm"].mean().plot.pie(autopct="%1.1f%%", figsize=(6, 6), ylabel="")
plt.title("Distribuição da Dependência Administrativa para o total Médio")
plt.show()
```

**obs:** Você pode fazer os gráficos considerando apenas os valores não nulos.
